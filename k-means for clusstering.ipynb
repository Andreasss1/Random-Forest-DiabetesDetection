```python
import pandas as pd
```

```python
df = pd.read_csv(r'D:\Downloads\DailyDelhiClimateTrain.csv')
df
```

```python
df.rename(columns={'wind_speed': 'flow'}, inplace=True)
df.rename(columns={'meanpressure': 'pressure'}, inplace=True)
df.rename(columns={'meantemp': 'temperature'}, inplace=True)
df
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Konversi kolom 'date' menjadi tipe data datetime
df['date'] = pd.to_datetime(df['date'])

# Pilih fitur untuk K-Means
features = ['flow', 'pressure', 'temperature', 'humidity']
X = df[features]

# Normalisasi data (opsional, tergantung pada karakteristik data)
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(X)

# Menentukan jumlah cluster (k) - disini k=3 sebagai contoh
k = 4

# Melakukan K-Means clustering
kmeans = KMeans(n_clusters=k, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Visualisasi hasil
plt.scatter(df['temperature'], df['humidity'], c=df['cluster'], cmap='viridis')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('K-Means Clustering')
plt.show()
```

```python
# Visualisasi hasil
plt.scatter(df['temperature'], df['flow'], c=df['cluster'], cmap='viridis')
plt.xlabel('Temperature')
plt.ylabel('Flow')
plt.title('K-Means Clustering')
plt.show()
```
