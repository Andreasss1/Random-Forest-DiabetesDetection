```python
import pandas as pd
```

```python
df = pd.read_csv(r'D:\Downloads\DailyDelhiClimateTrain.csv')
df
```

```python
df.rename(columns={'wind_speed': 'flow'}, inplace=True)
df.rename(columns={'meanpressure': 'pressure'}, inplace=True)
df.rename(columns={'meantemp': 'temperature'}, inplace=True)
df
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# Convert the date column to timestamp
df['date'] = pd.to_datetime(df['date'])

# Select relevant features
features = ['flow', 'temperature', 'pressure', 'humidity']
data = df[features]

# Initialize and train the Isolation Forest model
model = IsolationForest(contamination=0.1, random_state=42)
model.fit(data)

# Predict anomalies (1 for normal, -1 for anomalies)
df['anomaly'] = model.predict(data)

# Visualize anomalies
plt.scatter(df['date'], df['temperature'], c=df['anomaly'], cmap='viridis')
plt.xlabel('date')
plt.ylabel('flow')
plt.title('Isolation Forest Anomaly Detection')
plt.show()

```
